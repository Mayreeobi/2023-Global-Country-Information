```python
# Convert the datatypes of the Dataset

columns_to_convert = [
    "Density\n(P/Km2)",
    "Agricultural Land( %)",
    "Land Area(Km2)",
    "Armed Forces size",
    "Co2-Emissions",
    "CPI",
    "CPI Change (%)",
    "Forested Area (%)",
    "Gasoline Price",
    "GDP",
    "Gross primary education enrollment (%)",
    "Gross tertiary education enrollment (%)",
     "Minimum wage",
    "Out of pocket health expenditure",
    "Population",
    "Population: Labor force participation (%)",
    "Tax revenue (%)",
    "Total tax rate",
    "Unemployment rate",
    "Urban_population",
]
for col in columns_to_convert :
    df[col] = df[col].str.replace(r"[^0-9.]", "", regex=True).astype(float)
```

## GDP Per Capita

#### Find the average GDP per capita of all countries roundup to the nearest zeros

```python
# Calculate the average GDP per capita
average_gdp_per_capita = df['GDP per capita'].mean()

# Round to the nearest whole number
rounded_average_gdp = round(average_gdp_per_capita)

# Print the result
print(f"The average GDP per capita is: {rounded_average_gdp}")
```

#### Find the countries and in each continent that have the GDP per capita higher than the average GDP per capita?

```python
# Calculate the average GDP per capita
average_gdp_per_capita = df['GDP per capita'].mean()

# Filter countries with GDP per capita higher than the average for each continent
filtered_df = df[df['GDP per capita'] > average_gdp_per_capita]

# Display the result
print("Countries with GDP per capita higher than the average GDP per capita in each continent:")
for continent in df['Continent'].unique():
    continent_data = filtered_df[filtered_df['Continent'] == continent]
    if not continent_data.empty:
        print(f"\nContinent: {continent}")
        print(continent_data[['Country', 'GDP per capita']])
```

Insights:
    * The European country with GDP per capita higher than the average GDP per capita is Monaco with 184396.
    * The Asian country with GDP per capita higher than the average GDP per capita is Singapore with 65233.
    * In North America, the country with GDP per capita higher than the average GDP per capita is United States with 65280.68
    * In Oceania, the country with GDP per capita higher than the average GDP per capita is Australia with 54049.
    * In Africa, the only country with GDP per capita higher than the average GDP per capita is Seychelles with 17401.72
    * In South America, the countries with GDP per capita higher than the average GDP per capita is Venezuela with 16915.49

## Education

### How does the gross enrollment  for both primary and tertiary education vary by across continent

```python
columns_to_plot = ['Gross primary education enrollment (%)', 'Gross tertiary education enrollment (%)', 'Continent']

# Filter out rows with missing values in the specified columns
filtered_df = df[columns_to_plot].dropna()

# Create box plots using Seaborn
plt.figure(figsize=(12, 8))
sns.boxplot(data=filtered_df, x='Continent', y='Gross primary education enrollment (%)', palette='Blues', width=0.5)
plt.title('Gross Enrollment for Primary Education Across Continents')
plt.xlabel('Continent')
plt.ylabel('Gross primary education enrollment (%)')
plt.show()

plt.figure(figsize=(12, 8))
sns.boxplot(data=filtered_df, x='Continent', y='Gross tertiary education enrollment (%)', palette='Greens', width=0.5)
plt.title('Gross Enrollment for Tertiary Education Across Continents')
plt.xlabel('Continent')
plt.ylabel('Gross tertiary education enrollment (%)')
plt.show()
```

Insights:
    * There are outliers in both Primary and Teritary Education Enrollment.
    * In the Primary Education Enrollment, Africa have a larger viability than other continents. All continents are reasonably symmetric with the exception of Oceania which was left skewed.
    * Africa has the lowest teritary enrollment % while Europe has the highest percentage.
